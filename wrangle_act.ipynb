```python
#create copies
df_clean = df.copy()
df_predictions_clean = df_predictions.copy()
df_tweet_json_clean = df_tweet_json.copy()
```

```python
# Tidiness 1
# Define: In order to fulfill the tidiness criteria "each variable forms a column", 
# will melt the dog stages (doggo, floofer, pupper and puppo) into a single column called "dog_stage".

# Code
# handle none
df_clean.doggo.replace('None', '', inplace=True)
df_clean.floofer.replace('None', '', inplace=True)
df_clean.pupper.replace('None', '', inplace=True)
df_clean.puppo.replace('None', '', inplace=True)

# merge into column
df_clean['dog_stage'] = df_clean.doggo + df_clean.floofer + df_clean.pupper + df_clean.puppo

# handle multiple stages
df_clean.loc[df_clean.dog_stage == 'doggopupper', 'dog_stage'] = 'doggo, pupper'
df_clean.loc[df_clean.dog_stage == 'doggopuppo', 'dog_stage'] = 'doggo, puppo'
df_clean.loc[df_clean.dog_stage == 'doggofloofer', 'dog_stage'] = 'doggo, floofer'

# handle missing values
df_clean.loc[df_clean.dog_stage == '', 'dog_stage'] = np.nan

#Test
df_clean.dog_stage.head(10)
```

```python
# Quality 5
# Define: In df_predictions, there are 66 duplicated urls/dog pictures. Will remove duplicates 

# Code
df_predictions_clean = df_predictions_clean.drop_duplicates(subset='jpg_url', keep='first')

# Test
sum(df_predictions_clean.jpg_url.duplicated())
```

```python
# Quality 6
# Define: Condense image prediction into two columns, based on px_dog
# Code

dog_breed = []
confidence = []

def condense(df_predictions_clean):
    if df_predictions_clean.p1_dog == True:
        dog_breed.append(df_predictions_clean.p1)
        confidence.append(df_predictions_clean.p1_conf)
    elif df_predictions_clean.p2_dog == True:
        dog_breed.append(df_predictions_clean.p2)
        confidence.append(df_predictions_clean.p2_conf)
    elif df_predictions_clean.p3_dog == True:
        dog_breed.append(df_predictions_clean.p3)
        confidence.append(df_predictions_clean.p3_conf)
    else:
        dog_breed.append("not dog")
        confidence.append("not dog")

df_predictions_clean.apply(condense, axis = 1)
        
df_predictions_clean["dog_breed"] = dog_breed
df_predictions_clean["confidence"] = confidence

# Test
df_predictions_clean.head()
```

```python
# Quality 7
# Define: Remove not dog in dog_bread and confidence columns
# Code
df_predictions_clean = df_predictions_clean.drop(df_predictions_clean.query('dog_breed == "not dog"').index)
df_predictions_clean = df_predictions_clean.drop(df_predictions_clean.query('confidence == "not dog"').index)

# Test
df_predictions_clean.info()
```

```python
# Tidiness 2 (Part 2)
# Define: Drop unnecessary columns from df_predictions_clean. Combine with new df df_all_clean

#Code
df_predictions_clean = df_predictions_clean.drop(['jpg_url', 
                          'img_num', 
                          'p1', 
                          'p1_conf', 
                          'p1_dog', 
                          'p2', 
                          'p2_conf', 
                          'p2_dog', 
                          'p3', 
                          'p3_conf', 
                          'p3_dog'], axis = 1)


df_all_clean = df_all_clean.merge(df_predictions_clean, on = "tweet_id")

#Test
df_all_clean.head(1)
```

```python
#Quality 8
#Define: In df_all_clean, drop all denominator != 10
#Code
df_all_clean = df_all_clean.drop(df_all_clean.query('rating_denominator != "10"').index)

#Test
df_all_clean.info()
```
